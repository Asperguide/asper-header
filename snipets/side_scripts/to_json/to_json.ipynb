```python
# import the os module
import os
```

```python
# Specify typing in a retro-compatible manner
from typing import List, Union, Dict
```

```python
# import the json module
import json
```

```python
# Print local directory
print(f"cwd: {os.getcwd()}")
```

```python
# Load the file
with open("wtm.txt", "r", encoding="utf-8", newline="\n") as f:
    content: str = f.read()
    
print(f"Characters read: {len(content)}")
```

```python
# Convert to a array of line
document_lines: List[str] = content.split("\n")
print(f"Lines read: {len(document_lines)}")
```

```python
# Create json like dictionaries and add them to a list
key1:str = "fontName"
key2:str = "Logo"
final_list: List[Dict[str,Union[str, List[str]]]] = []
buffer: Dict[str, Union[str, List[str]]] = {}
for index, item in enumerate(document_lines):
    print(f"Iteration n°{index}")
    if "🚀" in item:
        print(f"Buffer state node: {buffer}")
        if len(buffer) > 0:
            final_list.append(buffer.copy())
            print("Buffer node added")
        buffer={}
        buffer[key1] = item.replace("🚀", "")
        buffer[key2] = []
    else:
        if isinstance(buffer[key2], List):
            buffer[key2].append(item)
        else:
            buffer[key2] = [item]
```

```python
# Diplay the final list
print(f"Final List length: {len(final_list)}")
print(f"Final List: {final_list}")
```

```python
# Jsonify the content
# if isinstance(final_list, List): # Use this piece of code if you wish to enforce a dictionnary, but it is not required for lists.
#     final_list = {"list":final_list}
if isinstance(final_list, (str, int, float)):
    final_list = {str(type(final_list)):final_list}
json_content:str = json.dumps(final_list, skipkeys=False, ensure_ascii=True, check_circular=True, indent=None, separators=(",",":"), sort_keys=True)
print(f"json content: {json_content}")
```

```python
# write json to file
with open("wtm.min.json","w",encoding="utf-8", newline="\n") as f:
    f.write(json_content)
print("write successfull.")
```
